```python
import pandas as pd
import numpy as np
```

```python
with open('./input/input4.txt') as f:
    lines = f.readlines()
    lines = [line.rstrip() for line in lines]
lines
```

```python
#part 1 answer
count = 0
for i in lines:
    list1 = i.split(',')[0]
    list2 = i.split(',')[1]
    range1 = list(range(int(list1.split('-')[0]),int(list1.split('-')[1])+1))
    range2 = list(range(int(list2.split('-')[0]),int(list2.split('-')[1])+1))
    if all(a in range1 for a in range2):
        count+=1
    elif all(a in range2 for a in range1):
        count+=1

    
count 
```

```python
#part 2 answer
count = 0
for i in lines:
    list1 = i.split(',')[0]
    list2 = i.split(',')[1]
    range1 = list(range(int(list1.split('-')[0]),int(list1.split('-')[1])+1))
    range2 = list(range(int(list2.split('-')[0]),int(list2.split('-')[1])+1))
    for a in range1:
        if a in range2:
            count+=1
            break
count
```
