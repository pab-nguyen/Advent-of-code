```python
with open('./input/input5.txt') as f:
    lines = f.readlines()
    lines = [line.rstrip() for line in lines]
lines
```

```python
stack = []
for i in lines[0:9]:
    stack.append([i[x:x+3] for x in range(0, len(i), 4)])
    
stack
```

```python
stacks = []
a = []
for i in stack[8]:
    for x in stack:
        try:
            a.append(x[int(i)-1])
        except:
            continue
    stacks.append(a)
    a = []

for i in stacks:
    i.remove(i[-1])
    for x in range(1,10):
        try:
            i.remove('   ')
        except:
            continue

stacks

```

```python
#part 1 answer
stacks1 = stacks
emptylist = False
for i in lines[10:]:
    count = 0
    for x in stacks1:
        try:
            count += len(x)
        except:
            count += len(x)
            continue
    a = i.split(" ")
    start = int(a[3])-1
    end = int(a[5])-1
    num = int(a[1])

    b = stacks1[start][0:num]
    
    if len(stacks1[end]) == 1 and stacks1[end][0] == 'Null':
        for x in b:
            stacks1[end].insert(0,x.rstrip())
            stacks1[end].pop(len(stacks1[end])-1)
    else:
        for x in b:
            stacks1[end].insert(0,x.rstrip())
    
    if len(b) == len(stacks1[start]):
        emptylist = True
    if emptylist == True:
        stacks1.pop(start)
    else:
        for x in range(0,num):
            stacks1[start].pop(0)
    if emptylist == True:
        stacks1.insert(start,[['Null']])
        emptylist = False
    
```

```python
string = ""
for i in stacks1:
    string += i[0][1]
string
```
